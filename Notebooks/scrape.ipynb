
Scrape www.filmtotaal.nl

Save to a series of files

stars;title;text

```python
import json
import pandas as pd
import re
import numpy as np
import os
import matplotlib.pyplot as plt
import datetime
from dateutil import tz
import urllib.request
from bs4 import BeautifulSoup
from dateutil import tz
```

## (Remove and re)make folder

```python
# Remove folder if exists
folder = 'data'

if os.path.exists(folder):
    os.remove(folder)

# Make folder again
os.makedirs(folder)
```

## Download

```python
for pagenum in range(255):

    url = 'https://www.filmtotaal.nl/recensie?page=' + str(pagenum)
    print(url)

    # Download and convert to list of strings
    page = urllib.request.urlopen(url)
    soup = BeautifulSoup(page, "html5lib")
    soup = str(soup).split('\n')


    # Strip start and end lines
    rgxstart = '<div class="layout-wrap">'
    rgxstop = 'start right sidebar'

    startline = []
    stopline = []
    for s in range(len(soup)):
        if not re.search(rgxstart, soup[s]) is None:
            startline.append(s)
        if not re.search(rgxstop, soup[s]) is None:
            stopline.append(s)

    if not len(startline) == 1:
        print ('ERROR: Startline not found.')
    elif len(stopline) == 0:
        print ('ERROR: Stopline not found.')
    else:
        soup = soup[startline[0]:stopline[0]]


    # Extract info
    rgxtitle = 'class="title"'
    rgxtxt = 'class="bite"'
    rgxstars = 'div class="stars'

    title = []
    txt = []
    stars = []

    for s in range(len(soup)):
        if not re.search(rgxtitle, soup[s]) is None:
            title_ = re.findall(r"<h2>.*</h2>", soup[s])[-1].strip('</?h2>')
            title.append(title_)
        elif not re.search(rgxstars, soup[s]) is None:
            stars_ = 0
            for i in range(s+1,s+7):
                if not re.search('whole',soup[i]) is None:
                    stars_ += 1
                elif not re.search('half',soup[i]) is None:
                    stars_ += 0.5
            stars.append(stars_)

        elif not re.search(rgxtxt, soup[s]) is None:
            txt_ = soup[s].strip(' ?(<p class="bite">|</p>)')
            txt.append(txt_)
            
            
    # Save to file
    if len(title) == len(stars) and len(title) == len(txt):
        res = pd.DataFrame(
            {'title': title,
             'stars': stars,
             'txt': txt
            }) 

        fileout = folder + '/' + re.sub('=', '-', url.split('recensie?')[1]) + '.csv'
        print(fileout)

        res.to_csv(fileout, sep=';', na_rep='', float_format=None, columns=None, header=True, index=False,\
                            index_label=None, mode='w', encoding=None, compression=None, quoting=None, quotechar='"',\
                            line_terminator='\n', chunksize=None, tupleize_cols=None, date_format=None, \
                            doublequote=True, escapechar=None, decimal='.')

    else:
        print ('ERROR: Unequal list lengths')

```
