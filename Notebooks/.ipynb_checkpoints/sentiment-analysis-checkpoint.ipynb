## Preprocess texts

In text analysis, we will often do some preprocessing.  The goal is 
- to remove unnecessary information, like punctiations
- to uniformise stuff that has the same meaning, e.g. "Hello, I am joris' means the same as "hello i am Joris"

````python
# Convert to lowercase
df.txt = [t.lower() for t in df.txt] 

# Remove punctuation 
df.txt = [re.sub('[.,;:?!]', '', t) for t in df.txt]

df[:10]
````

## Feature extraction: bag of words


Why?  

Texts are unstructured information.  Most ML algorithms need a matrix of features (numbers) as input.  Rows = instances, columns = features.

How?  

Make a list with all words in all texts = dictionary.  Count for each text the number of times each word occurs.  This is often zero.

-> **document-term matrix**

_Example_

    text0 'This is the first document.'
    text1 'This is the second second document.'
    text2 'And the third one.'
    text3 'Is this the first document?'
    
                and      document first   is      one     second  the     third  this
    text0       0        1        1       1       0       0       1       0      1
    text1       0        1        0       1       0       2       1       0      1
    text2       1        0        0       0       1       0       1       1      0
    text3       0        1        1       1       0       0       1       0      1



````python
# Bag of words
vectorizer = CountVectorizer()
counts = vectorizer.fit_transform(df.txt)
````

## Normalisation: tf-idf

Why?  Words that are common, like articles, have little differentiating value.

How?  Here, we use tf-idf (term-frequency times inverse document-frequency): 

\begin{equation}
tfidf (t,d,D) = tf (t,d) \cdot idf (t,D)
\end{equation}

$tf(t,d)$ are the term frequencies, i.e. the counts in the document-term matrix

$idf (t,D)$ is the inverse document frequency, a measure of how much information the word provides, that is, whether the word is common or rare across all documents:

\begin{equation}
    idf (t,D) = \frac{\log ⁡ N} { \{d ∈ D : t ∈ d\} }
\end{equation}
   

with

$N$ the total number of documents

$\{d ∈ D : t ∈ d\}$ the number of documents where the term $t$ appears

````python
transformer = TfidfTransformer(smooth_idf=False)
tfidf = transformer.fit_transform(counts)
````

## Train and test set


Why?  Train your model on a part of the data, test it on another part.  The most commom mistake for beginners in machine learning, is to test on train data.  Of course, you get good results if you do it like that.


How?  Simplest approach: split your data randomly in a train and a test set.  Larger projects: train, develop and test set.

<img src="../img/train-test.png" width="500">




````python
# Split in 75% training set en 25% test set
Xtrain, Xtest, ytrain, ytest = train_test_split(tfidf, df.stars, test_size=0.25, random_state=42)
````

## Nearest neighbour regression

For an unknown text: get stars from n most similar texts, a.k.a. the nearest neighbours and average.

Similar texts?  Vector distances.  In 2D (only two features), the Eulerian distance is inuitive:

<img src="../img/nearest-neighbours.png" width="500">



\begin{equation}
    d_{12} = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
\end{equation}

Other distance measures exist.

````python
# Set hyperparameters
n_neighbors = 10
weights = 'uniform'
````

````python
# Train model
knn = neighbors.KNeighborsRegressor(n_neighbors, weights=weights)
````

````python
# Predict
ypred = knn.fit(Xtrain, ytrain).predict(Xtest)
````

````python
# Inspect prediction
list(ypred[:10])
````

````python
# Compare to true values
list(ytest[:10])
````

## Evaluation

### Mean square error

Predictions for the star rating of the texts in the test set:

    ypred

True star ratings of the same texts:

    ytest
    
To campare these two, we calculate the average of the squared distance:
    
\begin{equation}
    MSE = \frac{1}{n} \sum_i {(ypred_i - ytest_i)^2}
\end{equation}


## Evaluation

### Root mean square error (RMSE)

Predictions for the star rating of the texts in the test set:

    ypred

True star ratings of the same texts:

    ytest
    
To campare these two, we calculate the average of the squared distance:
    
\begin{equation}
    MSE = \frac{1}{n} \sum_i {(ypred_i - ytest_i)^2}
\end{equation}

Take the root to get a value in the same units as target:

\begin{equation}
    RMSE = \sqrt{MSE}
\end{equation}



````python
# Mean square error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(ytest, ypred)

# Root mean square error
rmse = np.sqrt(mse)
print('RMSE = ' + str(rmse))
````

## Model improvement: hyperparameters

````python
# Fit regression model
n_neighbors = 5
weights = 'distance'

knn2 = neighbors.KNeighborsRegressor(n_neighbors, weights=weights)
ypred2 = knn2.fit(Xtrain, ytrain).predict(Xtest)

# RMSE
mse = mean_squared_error(ytest, ypred2)
rmse = np.sqrt(mse)
print('RMSE = ' + str(rmse))
````
