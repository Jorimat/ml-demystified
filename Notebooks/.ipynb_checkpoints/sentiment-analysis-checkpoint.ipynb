```python
import json
import pandas as pd
import re
import numpy as np
import os
# import stringdist
# import boto3 
# import botocore
# import csv

import matplotlib.pyplot as plt
from sklearn import neighbors
from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

# import operator
# import datetime
# from dateutil import tz
# from pyspark.sql.types import StringType
# from pyspark.sql.functions import udf
# import shutil

# import urllib.request
# from bs4 import BeautifulSoup
# from dateutil import tz
```

## List files in filder 'data'

```python
folder = 'data'
allfiles = os.listdir(folder)

# Pre-append folder to each item in list
allfiles = [folder + '/' + f for f in allfiles]

# Sort alphabetically
allfiles.sort()

# Show first ten
allfiles[:10]
```

Bekijk eens één of meerdere van deze files in een text editor en in een spreadsheetprogramma.

# Load files into pandas dataframe

```python
# Load first file
df = pd.read_csv(allfiles[0], sep=';', header='infer')

# Look at dataframe
df
```

```python
# Load all the other files and append to df
for f in range(1, len(allfiles)):
    df_ = pd.read_csv(allfiles[f], sep=';', header='infer')   
    df = df.append(df_)

# Remove duplicates in df
df = df.drop_duplicates()

# Reset index
df = df.reset_index()
```

```python
# Inspect
df
```

```python
# Count rows
len(df)
```

## Preprocess texts

```python
# Preprocess: to lower and remove dots etc
df.txt = [t.strip('.,?!;:"').lower() for t in df.txt] 

df[:10]
```

## Feature extraction: bag of words


Why: texts are unstructured information.  Most ML algorithms need a matrix of features (numbers) as input.

Rows = instances



How: make a list with all words in all texts = dictionary.  Count for each text the number of times each word occurs.  This is often zero.

# Bag of words
vectorizer = CountVectorizer()
counts = vectorizer.fit_transform(df.txt)

```python
# Inspect counts matrix
counts
```

```python
counts.shape
```

```python
counts.toarray()
```

```python
plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.spy(counts.toarray())
plt.show()
```

## Normalisation: tfidf

Waarom en formules
.............

```python
transformer = TfidfTransformer(smooth_idf=False)
tfidf = transformer.fit_transform(counts)
```

Doordenker: hoe ziet de figuur eruit als we dit runnen:

    plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
    plt.spy(tfidf.toarray())
    plt.show()

# What do we have now

figuurke met een grote matrix ......

features (weighted word counts) | targets (stars 0-5)

## Train and test set

.........


Waarom, meest gemaakte beginnersfout in ML, bij echte ontwikkeling train, develop and test set

```python
# Split in 75% training set en 25% test set
Xtrain, Xtest, ytrain, ytest = train_test_split(tfidf, df.stars, test_size=0.25, random_state=42)
```

```python
print(Xtrain.shape)
print(ytrain.shape)
print(Xtest.shape)
print(ytest.shape)
```

## Nearest neighbour regression

............



```python
# Set hyperparameters
n_neighbors = 5
weights = 'uniform'
```

```python
# Train model
knn = neighbors.KNeighborsRegressor(n_neighbors, weights=weights)
```

```python
# Predict
ypred = knn.fit(Xtrain, ytrain).predict(Xtest)
```

```python
# Inspect prediction
list(ypred[:10])
```

```python
# Compare to true values
list(ytest[:10])
```

## Evaluation: mean square error

.......

```python
# Mean square error
from sklearn.metrics import mean_squared_error
mean_squared_error(ytest, ypred)
```

## Model improvement: other hyper parameters

```python
# Fit regression model
n_neighbors = 20
weights = 'distance'

knn2 = neighbors.KNeighborsRegressor(n_neighbors, weights=weights)
ypred2 = knn.fit(Xtrain, ytrain).predict(Xtest)

# Mean square error
mean_squared_error(ytest, ypred)
```
